```python
import openai
from docx import Document
from groq import Groq
import os



api_key = "Replace with your own api key"
def generate_section_content(prompt):
    client = Groq(api_key=api_key)
    chat_completion = client.chat.completions.create(
        messages=[
            {
                "role": "user",
                "content": prompt
            }
        ],
        model="llama3-8b-8192",
    )
    return chat_completion.choices[0].message.content

def recitals(company_name, company_description, counter_party_name, counter_party_description, transaction_description):
    prompt = f"Write the 'Recitals' section of a Non-disclosure Agreement without an introductory phrase. \
    Include details about {company_name} which is {company_description}, {counter_party_name} which is {counter_party_description}, \
    and the transaction description which is {transaction_description}."
    return generate_section_content(prompt)

def purpose(agreement_purpose, transaction_description, recipient_party_description):
    prompt = f"Write the 'Purpose' section of a Non-disclosure Agreement without an introductory phrase. \
    Include the agreement purpose which is {agreement_purpose}, the transaction description which is {transaction_description}, \
    and the recipient party description which is {recipient_party_description}."
    return generate_section_content(prompt)

def definitions(affiliate_definition, confidential_info_definition):
    prompt = f"Write the 'Definitions' section of a Non-disclosure Agreement without an introductory phrase. \
    Include the definition of affiliate which is {affiliate_definition} and the definition of confidential information which is {confidential_info_definition}."
    return generate_section_content(prompt)

def disclosure_of_confidential_information(obligations_description, receipt_description, disclosure_description, third_party_disclosure_description, conditions_description, replication_restriction, ownership_rights, trademark_use, copying_restriction):
    prompt = f"Write the 'Disclosure of Confidential Information' section of a Non-disclosure Agreement without an introductory phrase. \
    Include obligations description which is {obligations_description}, receipt description which is {receipt_description}, disclosure description which is {disclosure_description}, \
    third party disclosure description which is {third_party_disclosure_description}, conditions description which is {conditions_description}, and counter party restrictions including replication restriction which is {replication_restriction}, ownership rights which are {ownership_rights}, trademark use which is {trademark_use}, and copying restriction which is {copying_restriction}."
    return generate_section_content(prompt)

def limitations_of_use_of_confidential_information(standard_of_care_description, retention_obligation):
    prompt = f"Write the 'Limitations of Use of Confidential Information' section of a Non-disclosure Agreement without an introductory phrase. \
    Include standard of care description which is {standard_of_care_description} and retention obligation which is {retention_obligation}."
    return generate_section_content(prompt)

def standard_of_care(pre_existing_information, public_knowledge, third_party_receipt, legal_disclosure):
    prompt = f"Write the 'Standard of Care' section of a Non-disclosure Agreement without an introductory phrase. \
    Include pre-existing information which is {pre_existing_information}, public knowledge which is {public_knowledge}, third party receipt which is {third_party_receipt}, \
    and legal disclosure which is {legal_disclosure}."
    return generate_section_content(prompt)

def disclosure_required_by_law_and_other_exceptions(contact_info_1, contact_info_2):
    prompt = f"Write the 'Disclosure Required by Law and Other Exceptions' section of a Non-disclosure Agreement without an introductory phrase. \
    Include contact info 1 which is {contact_info_1} and contact info 2 which is {contact_info_2}."
    return generate_section_content(prompt)

def contact_persons():
    prompt = "Write the 'Contact Persons' section of a Non-disclosure Agreement without an introductory phrase."
    return generate_section_content(prompt)

def term_and_termination(effective_date, post_termination_obligations, survival_period):
    prompt = f"Write the 'Term and Termination' section of a Non-disclosure Agreement without an introductory phrase. \
    Include effective date which is {effective_date}, post-termination obligations which are {post_termination_obligations}, and survival period which is {survival_period}."
    return generate_section_content(prompt)

def return_of_confidential_information(return_confidential_information, retention_exceptions, obligation_survival, list_retention, dispute_resolution):
    prompt = f"Write the 'Return of Confidential Information' section of a Non-disclosure Agreement without an introductory phrase. \
    Include return of confidential information which is {return_confidential_information}, retention exceptions which are {retention_exceptions}, \
    obligation survival which is {obligation_survival}, list retention which is {list_retention}, and dispute resolution which is {dispute_resolution}."
    return generate_section_content(prompt)

def warranty():
    prompt = "Write the 'Warranty' section of a Non-disclosure Agreement without an introductory phrase."
    return generate_section_content(prompt)

def breach(breach_responsibility, indemnification_provision, additional_remedies):
    prompt = f"Write the 'Breach' section of a Non-disclosure Agreement without an introductory phrase. \
    Include breach responsibility which is {breach_responsibility}, indemnification provision which is {indemnification_provision}, and additional remedies which are {additional_remedies}."
    return generate_section_content(prompt)

def applicable_law_consent_to_jurisdiction():
    prompt = "Write the 'Applicable Law; Consent to Jurisdiction' section of a Non-disclosure Agreement without an introductory phrase."
    return generate_section_content(prompt)

def dispute_resolution():
    prompt = "Write the 'Dispute Resolution' section of a Non-disclosure Agreement without an introductory phrase."
    return generate_section_content(prompt)

def miscellaneous():
    prompt = "Write the 'Miscellaneous' section of a Non-disclosure Agreement without an introductory phrase."
    return generate_section_content(prompt)

def signature_of_both_parties():
    prompt = "Write the 'Signature of Both Parties' section of a Non-disclosure Agreement without an introductory phrase."
    return generate_section_content(prompt)

def create_nda_document(inputs):
    doc = Document()
    doc.add_heading('Non-disclosure Agreement', level=1)

    # Add sections to the document
    sections = [
        ("Recitals", recitals, [inputs["company_name"], inputs["company_description"], inputs["counter_party_name"], inputs["counter_party_description"], inputs["transaction_description"]]),
        ("Purpose", purpose, [inputs["agreement_purpose"], inputs["transaction_description"], inputs["recipient_party_description"]]),
        ("Definitions", definitions, [inputs["affiliate_definition"], inputs["confidential_info_definition"]]),
        ("Disclosure of Confidential Information", disclosure_of_confidential_information, [inputs["obligations_description"], inputs["receipt_description"], inputs["disclosure_description"], inputs["third_party_disclosure_description"], inputs["conditions_description"], inputs["replication_restriction"], inputs["ownership_rights"], inputs["trademark_use"], inputs["copying_restriction"]]),
        ("Limitations of Use of Confidential Information", limitations_of_use_of_confidential_information, [inputs["standard_of_care_description"], inputs["retention_obligation"]]),
        ("Standard of Care", standard_of_care, [inputs["pre_existing_information"], inputs["public_knowledge"], inputs["third_party_receipt"], inputs["legal_disclosure"]]),
        ("Disclosure Required by Law and Other Exceptions", disclosure_required_by_law_and_other_exceptions, [inputs["contact_info_1"], inputs["contact_info_2"]]),
        ("Contact Persons", contact_persons, []),
        ("Term and Termination", term_and_termination, [inputs["effective_date"], inputs["post_termination_obligations"], inputs["survival_period"]]),
        ("Return of Confidential Information", return_of_confidential_information, [inputs["return_confidential_information"], inputs["retention_exceptions"], inputs["obligation_survival"], inputs["list_retention"], inputs["dispute_resolution"]]),
        ("Warranty", warranty, []),
        ("Breach", breach, [inputs["breach_responsibility"], inputs["indemnification_provision"], inputs["additional_remedies"]]),
        ("Applicable Law; Consent to Jurisdiction", applicable_law_consent_to_jurisdiction, []),
        ("Dispute Resolution", dispute_resolution, []),
        ("Miscellaneous", miscellaneous, []),
        ("Signature of Both Parties", signature_of_both_parties, [])
    ]

    for title, section_func, args in sections:
        doc.add_heading(title, level=2)
        content = section_func(*args)
        doc.add_paragraph(content)

    doc.save('NDA_Document.docx')
    print("NDA document has been created and saved as 'NDA_Document.docx'")

# Sample inputs dictionary
inputs = {
    "company_name": "ABC Corp",
    "company_description": "a technology company",
    "counter_party_name": "XYZ Inc",
    "counter_party_description": "a consulting firm",
    "transaction_description": "a business collaboration",
    "agreement_purpose": "to outline the terms of the business collaboration",
    "recipient_party_description": "the recipient party in this agreement",
    "affiliate_definition": "any entity that controls, is controlled by, or is under common control with a party",
    "confidential_info_definition": "all information disclosed by one party to the other that is marked as confidential",
    "obligations_description": "the obligations of the receiving party with respect to the confidential information",
    "receipt_description": "the manner in which the receiving party receives the confidential information",
    "disclosure_description": "the terms under which the confidential information may be disclosed",
    "third_party_disclosure_description": "conditions under which the confidential information can be disclosed to third parties",
    "conditions_description": "the conditions under which the confidential information must be kept",
    "replication_restriction": "restrictions on replicating the confidential information",
    "ownership_rights": "ownership rights related to the confidential information",
    "trademark_use": "use of trademarks associated with the confidential information",
    "copying_restriction": "restrictions on copying the confidential information",
    "standard_of_care_description": "the standard of care to be used in protecting the confidential information",
    "retention_obligation": "the obligation to retain the confidential information",
    "pre_existing_information": "information that pre-exists the agreement",
    "public_knowledge": "information that is already public knowledge",
    "third_party_receipt": "information received from third parties",
    "legal_disclosure": "disclosures required by law",
    "contact_info_1": "contact@abc.com",
    "contact_info_2": "contact@xyz.com",
    "effective_date": "2024-08-01",
    "post_termination_obligations": "obligations that continue after termination of the agreement",
    "survival_period": "the period during which obligations survive termination",
    "return_confidential_information": "the return of confidential information after termination",
    "retention_exceptions": "exceptions to the retention of confidential information",
    "obligation_survival": "survival of obligations related to confidential information",
    "list_retention": "retention of lists and summaries",
    "dispute_resolution": "mechanisms for resolving disputes",
    "breach_responsibility": "responsibility for breach of the agreement",
    "indemnification_provision": "indemnification provisions",
    "additional_remedies": "additional remedies for breach"
}

# Run the function to create the document
create_nda_document(inputs)
```
